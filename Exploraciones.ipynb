```python
mRatings['genre']=mRatings[['Action','Animation','Comedy','Drama','Documentary','Romance']].idxmax(1) 
```

```python
mRatings['genre']
```

```python
genres_map = {"Action": 'red', "Animation": 'blue', "Comedy": 'yellow', "Drama": 'green', "Documentary": 'orange', "Romance": 'purple'}
mRatings['genre_color']=mRatings['genre'].apply(lambda x: genres_map[x])
```

```python
mRatings['genre_color']
```

```python
pd.pivot_table(mRatings,values='rating',index='year',columns=['genre'],aggfunc=np.mean)
```

```python
plot1 = mRatings.groupby('genre').length.plot(kind='kde',
                                              xlim=(0,120),
                                              legend='genre'
                                             )

plot1[0].set_xlabel('Number of Minutes')
plot1[0].set_title('Distribution of Films by Runtime Minutes')
```

```python
plot2 = mRatings.groupby('genre').rating.plot(kind='kde',
                                              xlim=(0,10),
                                              legend='genre'
                                             )
plot2[0].set_xlabel('Ratings')
plot2[0].set_title('Distribution of Ratings')
```

```python
pd.pivot_table(mRatings,values='rating',
               index='title',
               columns=['genre']).plot(kind='box',by='genre').\
set_ylabel('Rating')
```

```python
pd.plotting.scatter_matrix(mRatings[['year','length','budget','rating','votes']], 
                           alpha=0.2, 
                           figsize=(6, 6), 
                           diagonal='kde'
                          )
plt.show()        
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15,15))
row = 0
col = 0
for index, genre in mRatings.groupby('genre'):
    if row > 2:
        row = 0
        col += 1
    genre.groupby('genre').\
plot(ax=axes[row,col],kind='scatter',x='length',y='rating',s=np.sqrt(genre['votes']),c=genre['genre_color'],xlim=(0,120),ylim=(0,10),alpha=0.5,label=index)
    row += 1
   
```

```python
%matplotlib inline

import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
oil = pd.read_csv('https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/oil.csv')
```

```python
oil.tail()
```

```python
oil = oil[~np.isnan(oil[oil.columns[1]])]
```

```python
oil.tail()
```

```python
oil.dtypes
```

```python
oil['Year'] = pd.to_datetime(oil['Year'])
oil 
```

```python
oil.rename(columns = {oil.columns[1]: 'Oil_Price_1997_Dollars'},
           inplace=True
          )
```

```python
oil.head()
```

```python
matplotlib.style.use('ggplot')
```

```python
oil.plot(x='Year',
         y='Oil_Price_1997_Dollars'
        )
```

```python
from pandas.plotting import lag_plot

lag_plot(oil.Oil_Price_1997_Dollars)
```

```python
from pandas.plotting import autocorrelation_plot

autocorrelation_plot(oil['Oil_Price_1997_Dollars'])
```

```python
car_crashes=pd.read_csv("car_crashes.csv")
car_crashes.Year=car_crashes.Year.astype(str)
car_crashes
```

```python
car_crashes['Year']=pd.to_datetime(car_crashes['Year'])
car_crashes
```

```python
car_crashes.head()
```

```python
car_crashes['Car_Crash_Fatalities_US']=car_crashes['Car_Crash_Fatalities_US']/1000
car_crashes
#oil
```

```python
oil_prices_car_crashes = pd.merge(oil,
                                  car_crashes,
                                  #right_on=oil['Year']
                                 )
oil_prices_car_crashes
```

```python
oil_prices_car_crashes.plot(x='Year')
```

```python
autocorrelation_plot(oil_prices_car_crashes[['Car_Crash_Fatalities_US',
                                             'Oil_Price_1997_Dollars']
                                           ])
```
